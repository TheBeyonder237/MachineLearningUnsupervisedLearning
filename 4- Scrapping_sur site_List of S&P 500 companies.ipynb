```python
import yfinance as yf
ticker = yf.Ticker("AAPL")
```

```python
data = yf.download(tickers = 'AAPL', start="2018-01-01", end="2023-05-29")
```

```python
data
```

```python
data.columns
```

```python
# Remettre la colone d'index
dataset = data.reset_index()
dataset.head()
```

```python
dataset = dataset[["Date", "Adj Close"]]
dataset
```

```python
dataset = dataset.rename(columns = {"Date":"ds", "Adj Close":"y"})
dataset.head()
```

```python
from prophet import Prophet
```

```python
pht = Prophet(daily_seasonality = True)
pht.fit(dataset)
```

```python
future = pht.make_future_dataframe(periods = 365)
prediction = pht.predict(future)
fig = pht.plot(prediction)
plt.title("Prediction  of bitcoin Stock price")
plt.xlabel("Data")
plt.ylabel("Close Stock price")
```
